```python
s = list(input().split())
loss_list = []
for i,t in enumerate(s):
    if t == 'Train':
        loss_list.append(float(s[i+4]))
print(loss_list)

import matplotlib.pyplot as plt

x_values = list(range(len(loss_list)))
plt.plot(x_values, loss_list)
```
